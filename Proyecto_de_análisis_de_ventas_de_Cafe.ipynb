# Proyecto de análisis de ventas de Cafe

Este proyecto comienza con la carga de datos de ventas de café desde un archivo CSV llamado `Coffe_sales.csv`. Para ello, se utiliza la librería `pandas` de Python, para el manejo y análisis de datos.

El código inicial se encarga de lo siguiente:

1.  **Subir el archivo:** Se utiliza la función `files.upload()` de `google.colab` para permitir al usuario subir el archivo CSV directamente en el entorno de Colab.
2.  **Confirmar la carga:** Se muestra un mensaje indicando el nombre y tamaño del archivo subido.
3.  **Leer el archivo con pandas:** Se lee el contenido del archivo CSV y se carga en un DataFrame de pandas llamado `df`. Un DataFrame es una estructura de datos tabular, similar a una hoja de cálculo, que facilita la manipulación de los datos.
4.  **Mostrar una vista previa:** Se muestran las primeras filas del DataFrame (`df.head()`) para tener una idea rápida de la estructura y el contenido de los datos cargados.
5.  **Manejo de errores:** Se incluye un bloque `try-except` para capturar posibles errores durante la lectura del archivo, lo que hace que el código sea más robusto.

Una vez que los datos están cargados en el DataFrame `df`, se pueden realizar diversas operaciones de análisis, limpieza y visualización para obtener información valiosa sobre las ventas de café.

```python
from google.colab import files
import pandas as pd

uploaded = files.upload()

for fn in uploaded.keys():
  print('El usuario subió el archivo "{name}" con tamaño {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
  try:
    # Leer archivo con pandas
    df = pd.read_csv(fn)
    print("\nContenido del archivo:")
    display(df.head())
  except Exception as e:
    print(f"Error al leer el archivo: {e}")
```

```python
# Mostrar estadísticas descriptivas del DataFrame
display(df.describe())
```

**Interpretación de las Estadísticas Descriptivas:**

Observando la tabla generada con `.describe()`:

*   **count**: Nos dice cuántas filas tienen valores no nulos para cada columna. En este caso, todas las columnas numéricas tienen 3547 valores, lo que significa que no hay valores faltantes en estas columnas.
*   **mean (media)**: Es el promedio de los valores en cada columna. Por ejemplo, la `hour_of_day` promedio es aproximadamente 14.19 (alrededor de las 2 PM) y el `money` (dinero/venta) promedio es de aproximadamente 31.65.
*   **std (desviación estándar)**: Mide la dispersión de los datos alrededor de la media. Una desviación estándar alta indica que los valores están más dispersos. Por ejemplo, la desviación estándar de `hour_of_day` es de aproximadamente 4.23, lo que sugiere que hay una variación considerable en las horas en que se realizan las ventas.
*   **min (valor mínimo)**: El valor más bajo en cada columna. La hora más temprana de venta es la hora 6 (6 AM) y la venta mínima es de 18.12.
*   **25% (primer cuartil)**: El valor por debajo del cual se encuentra el 25% de los datos. El 25% de las ventas se realizaron antes de la hora 10 (10 AM) y el 25% de las ventas fueron menores a 27.92.
*   **50% (mediana)**: El valor central de los datos cuando están ordenados. Es el segundo cuartil. La mediana de `hour_of_day` es 14 (2 PM) y la mediana de `money` es 32.82. La mediana y la media son relativamente cercanas para `hour_of_day` y `money`, lo que podría indicar una distribución de datos algo simétrica para estas columnas.
*   **75% (tercer cuartil)**: El valor por debajo del cual se encuentra el 75% de los datos. El 75% de las ventas se realizaron antes de la hora 18 (6 PM) y el 75% de las ventas fueron menores a 35.76.
*   **max (valor máximo)**: El valor más alto en cada columna. La hora más tardía de venta es la hora 22 (10 PM) y la venta máxima es de 38.70.

En resumen, estas estadísticas nos dan una idea general de la distribución y el rango de valores en las columnas numéricas de tu conjunto de datos de ventas de café.

## Exploración de datos categóricos

### Subtask:
Analizar las columnas `cash_type`, `coffee_name`, `Time_of_Day`, `Weekday`, y `Month_name` para entender la distribución de ventas por tipo de pago, nombre del café, momento del día, día de la semana y mes.


**Reasoning**:
Calculate and display the value counts for each of the specified categorical columns to understand their distributions.



```python
# Calcular y mostrar la frecuencia de cada categoría para cash_type
print("Distribución de ventas por tipo de pago:")
display(df['cash_type'].value_counts())

# Calcular y mostrar la frecuencia de cada categoría para coffee_name
print("\nDistribución de ventas por nombre del café:")
display(df['coffee_name'].value_counts())

# Calcular y mostrar la frecuencia de cada categoría para Time_of_Day
print("\nDistribución de ventas por momento del día:")
display(df['Time_of_Day'].value_counts())

# Calcular y mostrar la frecuencia de cada categoría para Weekday
print("\nDistribución de ventas por día de la semana:")
display(df['Weekday'].value_counts())

# Calcular y mostrar la frecuencia de cada categoría para Month_name
print("\nDistribución de ventas por mes:")
display(df['Month_name'].value_counts())
```

## Análisis de ventas por tiempo

### Subtask:
Investigar la relación entre las ventas (`money`) y las columnas relacionadas con el tiempo (`hour_of_day`, `Time_of_Day`, `Weekday`, `Month_name`, `Date`, `Time`). Esto podría incluir analizar las ventas por hora del día, día de la semana o mes.


**Reasoning**:
Calcular las ventas promedio por hora del día, momento del día, día de la semana y mes para investigar la relación entre las ventas y las columnas relacionadas con el tiempo.



```python
# Calcular las ventas promedio por hora del día
ventas_por_hora = df.groupby('hour_of_day')['money'].mean()

# Calcular las ventas promedio por momento del día
ventas_por_momento_del_dia = df.groupby('Time_of_Day')['money'].mean()

# Calcular las ventas promedio por día de la semana
ventas_por_dia_de_la_semana = df.groupby('Weekday')['money'].mean()

# Calcular las ventas promedio por mes
ventas_por_mes = df.groupby('Month_name')['money'].mean()

# Mostrar los resultados
print("Ventas promedio por hora del día:")
display(ventas_por_hora)

print("\nVentas promedio por momento del día:")
display(ventas_por_momento_del_dia)

print("\nVentas promedio por día de la semana:")
display(ventas_por_dia_de_la_semana)

print("\nVentas promedio por mes:")
display(ventas_por_mes)
```

## Análisis de ventas por tipo de café y tipo de pago

### Subtask:
Examinar cómo varían las ventas según el tipo de café (`coffee_name`) y el tipo de pago (`cash_type`).


**Reasoning**:
Calculate and display the average sales for each coffee type and cash type.



```python
# Calcular las ventas promedio por tipo de café
ventas_por_tipo_cafe = df.groupby('coffee_name')['money'].mean()

# Calcular las ventas promedio por tipo de pago
ventas_por_tipo_pago = df.groupby('cash_type')['money'].mean()

# Mostrar las ventas promedio por tipo de café
print("Ventas promedio por tipo de café:")
display(ventas_por_tipo_cafe)

# Mostrar las ventas promedio por tipo de pago
print("\nVentas promedio por tipo de pago:")
display(ventas_por_tipo_pago)
```

## Identificación de tendencias y patrones

### Subtask:
Buscar tendencias a lo largo del tiempo o patrones en las ventas que puedan ser útiles para la toma de decisiones.


**Reasoning**:
Convert the 'Date' column to datetime, group by 'Date' to sum sales, group by 'Month_name' and 'Weekday' to analyze sales trends, and display the results.



```python
# Convertir la columna 'Date' a tipo datetime
df['Date'] = pd.to_datetime(df['Date'])

# Agrupar por 'Date' y calcular la suma total de ventas por día
ventas_totales_por_dia = df.groupby('Date')['money'].sum()

# Agrupar por 'Month_name' y 'Weekday' para analizar la tendencia de ventas por día de la semana dentro de cada mes
ventas_por_mes_y_dia = df.groupby(['Month_name', 'Weekday'])['money'].mean()

# Mostrar los resultados
print("Suma total de ventas por día:")
display(ventas_totales_por_dia)

print("\nVentas promedio por mes y día de la semana:")
display(ventas_por_mes_y_dia)
```

## Visualización de datos

### Subtask:
Crear gráficos y visualizaciones para representar los hallazgos de los análisis anteriores de manera clara y comprensible.


**Reasoning**:
Import the necessary library and create the requested visualizations based on the previously calculated data.



```python
import matplotlib.pyplot as plt

# Configurar el tamaño de las figuras para una mejor visualización
plt.figure(figsize=(10, 6))
# Crear un gráfico de barras de las ventas promedio por hora del día
plt.bar(ventas_por_hora.index, ventas_por_hora.values)
plt.xlabel("Hora del Día")
plt.ylabel("Ventas Promedio")
plt.title("Ventas Promedio por Hora del Día")
plt.xticks(ventas_por_hora.index) # Asegurar que se muestren todas las horas
plt.grid(axis='y', linestyle='--') # Añadir cuadrícula para facilitar la lectura
plt.show()

plt.figure(figsize=(8, 5))
# Crear un gráfico de barras de las ventas promedio por momento del día
plt.bar(ventas_por_momento_del_dia.index, ventas_por_momento_del_dia.values, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel("Momento del Día")
plt.ylabel("Ventas Promedio")
plt.title("Ventas Promedio por Momento del Día")
plt.grid(axis='y', linestyle='--')
plt.show()

plt.figure(figsize=(10, 6))
# Crear un gráfico de barras de las ventas promedio por día de la semana
plt.bar(ventas_por_dia_de_la_semana.index, ventas_por_dia_de_la_semana.values, color='salmon')
plt.xlabel("Día de la Semana")
plt.ylabel("Ventas Promedio")
plt.title("Ventas Promedio por Día de la Semana")
plt.grid(axis='y', linestyle='--')
plt.show()

plt.figure(figsize=(12, 6))
# Crear un gráfico de barras de las ventas promedio por mes
plt.bar(ventas_por_mes.index, ventas_por_mes.values, color='teal')
plt.xlabel("Mes")
plt.ylabel("Ventas Promedio")
plt.title("Ventas Promedio por Mes")
plt.grid(axis='y', linestyle='--')
plt.show()

plt.figure(figsize=(14, 7))
# Crear un gráfico de barras de las ventas promedio por tipo de café
plt.bar(ventas_por_tipo_cafe.index, ventas_por_tipo_cafe.values, color='purple')
plt.xlabel("Tipo de Café")
plt.ylabel("Ventas Promedio")
plt.title("Ventas Promedio por Tipo de Café")
plt.xticks(rotation=45, ha='right') # Rotar etiquetas del eje x
plt.grid(axis='y', linestyle='--')
plt.tight_layout() # Ajustar el layout para evitar solapamiento
plt.show()

plt.figure(figsize=(15, 7))
# Crear un gráfico de líneas de las ventas totales por día
plt.plot(ventas_totales_por_dia.index, ventas_totales_por_dia.values, marker='o', linestyle='-')
plt.xlabel("Fecha")
plt.ylabel("Ventas Totales")
plt.title("Ventas Totales por Día")
plt.xticks(rotation=45) # Rotar etiquetas del eje x
plt.grid(True)
plt.tight_layout()
plt.show()
```
